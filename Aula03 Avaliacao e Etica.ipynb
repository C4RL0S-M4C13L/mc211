## Exemplos de Condução Científica Errada

## In 2016 the scientific publisher Springer Nature retracted 58 papers from seven journals, authored mostly by Iran-based researchers, because the papers showed evidence of authorship manipulation, peer-review manipulation, and/or plagiarism.

## Ohio University in 2006 alleged more than three dozen cases of plagiarism in master's degree theses dating back 20 years in its mechanical engineering department. 

---

## A former faculty member involved in the plagiarism cases, **Jay S. Gunasekera**, was removed from his position as department chair, had his title of "distinguished professor" rescinded.

## 486 Chinese cancer researchers were found guilty of engaging in a fraudulent peer-review scheme by China's Ministry of Science and Technology. The investigation was initiated after the retraction of 107 papers published in Tumor Biology between 2012 and 2016. This is reported to be the most papers retracted from one journal.

## Trends in Retraction Rates

### The number of retractions in scientific publications has been increasing over time:

## - Before 2000, there were fewer than 100 annual retractions
## - By 2014, this number had grown to nearly 1000 retractions per year
## - In 2023, over 10,000 research papers were retracted globally, marking a new record

## However, the rate of increase in retractions has slowed in recent years:

## - The retraction rate roughly doubled from 2003 to 2009
## - Since 2012, the rate has remained relatively level
## - Retractions still remain rare, with only about 4 out of every 10,000 papers being retracted

## Reasons for Retraction

## The **Retraction Watch** database provides insights into reasons for retraction:

## - Nearly 40% of retraction notices did not mention fraud or misconduct
## - About 50% of retractions involved fabrication, falsification, or plagiarism
## - 10% were due to behaviors like forged authorship or fake peer reviews

## The COPE Report 2000 Annual Report Committee on Publication Ethics
### Edited Caroline White - Freelance Journalist

## **Má conduta grave em pesquisa**
### * Fabricação: invenção de dados ou casos
### * Falsificação: distorção intencional de dados
### * Plágio: cópia de ideias, dados ou palavras sem atribuição
### * Não obter consentimento de um comitê de ética para pesquisa
### * Não admitir que alguns dados estão faltando
### * Ignorar outliers sem declarar
### * Não incluir dados sobre efeitos colaterais em um ensaio clínico
### * Conduzir pesquisa em humanos sem consentimento informado ou sem justificar por que o consentimento não foi obtido para um comitê de ética
### * Publicação de análises *post hoc* sem declaração de que foram *post hoc*
### * Presentar autoria
### * Não atribuir outros autores
### * Publicação redundante
### * Não divulgar um conflito de interesses
### * Não tentar publicar pesquisa concluída
### * Não fazer uma busca adequada de pesquisa existente antes de começar uma nova pesquisa 
## **Má conduta menos grave**

## Ética e Moral

- **Ética** é a área da filosofia dedicada às ações e ao comportamento humano, filosofia moral. O objeto de estudo da ética são os princípios que orientam as ações humanas e a capacidade de avaliar essas ações.

- **Ética** e **moral** se diferenciam por a ética ser compreendida de maneira universal, enquanto a moral está sempre ligada aos fatores sociais e culturais que influenciam os comportamentos.

De uma forma sucinta, a ética é uma teoria que se ocupa dos princípios que orientam as ações, já a moral é prática e está relacionada às regras de conduta.
