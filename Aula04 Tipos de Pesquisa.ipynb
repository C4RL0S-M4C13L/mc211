## 1. Por que isso importa?

- Objetivo da aula  
  - Entender os principais tipos de pesquisa usados em engenharia e quando usar cada um.  
  - Relacionar tipo de pesquisa com pergunta científica e com o projeto/TCC.  
- Pergunta inicial para a turma  
  - “Quando vocês pensam em ‘pesquisa em engenharia’, o que imaginam: experimento de laboratório, simulação, questionário, modelagem matemática, estudo de caso na indústria?”

## 2. Classificações gerais de pesquisa

### 2.1 Quanto aos objetivos

- Exploratória  
  - Quando o tema é pouco conhecido, busca levantar possibilidades, hipóteses, dimensões do problema.  
- Descritiva  
  - Foca em caracterizar fenômenos (como algo é, como se distribui, quais perfis), sem necessariamente testar causa.  
- Explicativa  
  - Procura compreender relações de causa/efeito ou mecanismos, testar hipóteses.  

### 2.2 Quanto à natureza

- Pesquisa básica x aplicada  
  - Básica: foco em ampliar conhecimento teórico, sem aplicação imediata.  
  - Aplicada: voltada para solução de problemas concretos (processos, produtos, sistemas).  

### 2.3 Quanto à abordagem

- Quantitativa  
  - Trabalha com mensuração, dados numéricos, estatística, experimentos, simulações, modelos computacionais.  
- Qualitativa  
  - Trabalha com compreensão profunda de contextos, entrevistas, observação, estudos de caso com foco em significados.  
- Mista  
  - Combina abordagens quantitativa e qualitativa de forma planejada.

## 3. Estratégias de pesquisa típicas em engenharia

### 3.1 Pesquisa bibliográfica e documental

- Pesquisa bibliográfica  
  - Focada em revisar criticamente o conhecimento já publicado (livros, artigos, teses).  
- Pesquisa documental  
  - Usa documentos “brutos”: normas, relatórios técnicos, manuais, bases de dados internos, logs de sistema.  

### 3.2 Estudo de caso

- Características  
  - Investigação aprofundada de um caso (uma empresa, uma planta, uma linha de produção, um hospital, um sistema).  
  - Pode ser único ou múltiplo; pode combinar dados qualitativos e quantitativos.  
- Quando usar  
  - Para compreender em detalhe a implementação de uma solução, barreiras, resultados em contexto real.  

### 3.3 Pesquisa experimental

- Características  
  - Manipulação deliberada de uma ou mais variáveis independentes e observação de efeitos em variáveis dependentes, com controle de outras condições.  
- Exemplos em engenharia  
  - Ensaios de materiais, testes de protótipos em bancada, experimentos de controle de processos.  
- Noções básicas  
  - Conceitos de grupo controle/experimental, repetição, aleatorização (quando possível), planejamento fatorial (menção simples).  

### 3.4 Pesquisa por simulação e modelagem

- Simulação computacional  
  - Uso de modelos (discretos, contínuos, Monte Carlo, CFD, circuitos, etc.) para estudar o comportamento de sistemas sob cenários variados.  
- Modelagem matemática/computacional  
  - Formulação de equações, redes, algoritmos e modelos (por exemplo, modelos probabilísticos, modelos causais) que representam o sistema.  
- Quando usar  
  - Quando experimentos reais são caros, arriscados ou inviáveis; para explorar cenários e sensibilidades.  

### 3.5 Pesquisa de levantamento (survey)

- Características  
  - Uso de questionários ou formulários estruturados para coletar dados de uma amostra de indivíduos/organizações.  
- Exemplos  
  - Levantamento de percepção de risco, adoção de tecnologia, satisfação com um sistema, práticas de manutenção.  
- Pontos de atenção  
  - Amostragem, tamanho de amostra, viés de resposta, qualidade das perguntas.  

### 3.6 Pesquisa-ação / pesquisa participativa (opcional)

- Conceito  
  - Pesquisa em que o pesquisador atua conjuntamente com os participantes para mudar uma realidade, enquanto a estuda.  
- Exemplos  
  - Implementar melhorias em um processo fabril junto com a equipe de chão de fábrica, avaliando iterativamente resultados.

## 4. Conexão com a pergunta de pesquisa e o TCC

### 4.1 Como escolher o tipo de pesquisa

- Critério 1: Tipo de pergunta  
  - “Como é / quanto tem / com que frequência?” → tende a ser descritiva, survey, análise de dados existentes.  
  - “O que acontece se…?” / “Qual efeito de X em Y?” → experimento ou simulação.  
  - “Como se implementa / por que deu certo/errado?” → estudo de caso, pesquisa-ação.  
- Critério 2: Recursos e contexto  
  - Acesso a laboratório, dados, empresas, tempo de curso, competências da equipe.  

### 4.2 Mini-atividade em sala

- Atividade rápida (10–15 minutos)  
  - Em grupos, cada aluno escolhe um tema de interesse para TCC (ou um exemplo dado por você).  
  - Tarefa:  
    - Formular, em 1–2 frases, a pergunta de pesquisa.  
    - Indicar: objetivo (exploratório/descritivo/explicativo), natureza (básica/aplicada), abordagem (quantitativa/qualitativa/mista).  
    - Sugerir 1 estratégia principal: experimento, simulação, estudo de caso, survey, modelagem etc.  
- Discussão plenária  
  - Alguns grupos apresentam; você comenta coerência entre pergunta e tipo de pesquisa.

## 5. Fechamento da aula

- “Não existe ‘melhor’ tipo de pesquisa em geral; existe o tipo mais adequado à pergunta, ao contexto e aos recursos.”  

- “Em engenharia, experimentos, simulações, estudos de caso e levantamentos são ferramentas diferentes para responder perguntas diferentes.”
