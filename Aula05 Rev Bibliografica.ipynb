## 1. Provocação

- Pergunta:
  - "Quem aqui já começou um projeto sem ler nada antes?"
  - "O que vocês acham que é mais 'científico': propor algo totalmente 'original' do zero ou construir sobre o que já foi feito?"
- Mostre a ideia central:
  - Nenhuma pesquisa começa do zero; toda pesquisa é um diálogo com trabalhos anteriores, e a revisão é o lugar em que esse diálogo acontece.
  - Sem revisão, o projeto é apenas opinião; com revisão, ganha base teórica, contexto histórico e método ancorado em evidência.

***

## 2. O que é revisão bibliográfica 

- Definição:
  - Revisão bibliográfica é a etapa em que o pesquisador localiza, seleciona, analisa criticamente e integra o conhecimento já publicado sobre um tema.
  - Trabalha com fontes secundárias (artigos, livros, teses etc.), diferindo da coleta de dados empíricos.
- Não é:
  - Lista descritiva de resumos.
  - Colagem de citações.
- É:
  - Análise crítica, síntese e organização do campo: quem são os autores-chave, quais conceitos, quais métodos, quais resultados e controvérsias.

***

## 3. Por que fazer revisão bibliográfica? 

1. **Evitar reinvenção e redundância**
   - Evita repetir estudos já feitos, desperdiçando tempo, recurso e esforço.
   - Ajuda a identificar soluções existentes, limitações e o que de fato ainda falta investigar.

2. **Identificar lacunas e formular boas perguntas**
   - A revisão permite ver onde há consenso, onde há conflito e onde quase não há estudos.
   - As lacunas identificadas justificam a relevância da sua pesquisa: "por que meu trabalho é necessário?".

3. **Construir embasamento teórico e conceitual sólido**
   - Garante que o trabalho se apoie em teorias, modelos e evidências reconhecidas na área.
   - Ajuda a definir conceitos com precisão (por exemplo, diferentes noções de "causalidade", "probabilidade", "risco" etc.).

4. **Apoiar escolhas metodológicas**
   - Mostra quais métodos têm sido usados, com quais resultados e limitações.
   - Ajuda a justificar o delineamento, o tipo de estudo, as métricas e as técnicas de análise escolhidas.

5. **Situar o trabalho no "estado da arte"**
   - A revisão mostra como seu estudo se conecta à linha de desenvolvimento de um campo.
   - Isso é crucial para artigos, dissertações e teses, onde avaliadores querem ver se o estudo dialoga com a fronteira do conhecimento.

6. **Dar credibilidade e qualidade científica**
   - Uma revisão consistente aumenta a confiança de leitores, banca e pareceristas na seriedade do trabalho.
   - Revisões de boa qualidade são vistas como um indicador de maturidade acadêmica do autor.

***

## 4. Por que isso importa para o aluno (e não só "para a ciência")

- **Melhorar argumentação no texto**
  - Facilita escrever introdução, justificativa e discussão, pois o aluno tem "material" para comparar e dialogar.
- **Facilitar publicação e defesa**
  - Trabalhos bem ancorados em literatura têm maior chance de aceitação em periódicos e de boa avaliação em bancas.
- **Otimizar tempo**
  - Uma boa revisão no início evita retrabalho metodológico e mudanças bruscas de tema no meio do caminho.

***

## 5. Como fazer uma revisão bibliográfica (Passo a Passo)

### 5.1 Definir escopo e pergunta

- **Tema**: Qual é o assunto geral que você vai revisar?
- **Pergunta ou objetivo**: Que pergunta específica guia a revisão? (ex.: "Quais métodos causais têm sido usados em diagnóstico clínico?")
- **Recorte temporal**: Que período vai cobrir? (últimos 5, 10, 20 anos?)
- **Área e tipo de estudo**: Vai revisar só artigos peer-reviewed, inclui teses, livros, preprints?
- **Idioma**: Português, inglês, ou ambos?

### 5.2 Planejar e executar a busca

- **Escolha as bases de dados adequadas**:
  - Scopus e Web of Science: multidisciplinares, abrangentes, com citações.
  - IEEE Xplore: engenharia e computação.
  - PubMed: biomedicina e saúde.
  - SciELO: pesquisa latino-americana.
  - Google Scholar: busca rápida, mas menos rigorosa.
  - Repositórios de teses (CAPES, institucionais).

- **Defina palavras-chave e operadores booleanos**:
  - Use AND para combinar conceitos: "Bayesian networks AND diagnosis".
  - Use OR para alternativas: "causal inference OR causal discovery".
  - Use NOT para excluir: "Bayesian NOT neural networks".
  - Teste buscas preliminares para ajustar termos muito amplos ou muito estreitos.

- **Registre a estratégia de busca** (bases consultadas, datas, termos, filtros aplicados) para garantir transparência e reprodutibilidade.

### 5.3 Selecionar estudos (triagem)

- **Primeira triagem** (título e resumo):
  - Aplique rapidamente os critérios de inclusão/exclusão.
  - Descarte o que é claramente irrelevante.

- **Segunda triagem** (leitura completa):
  - Leia o texto completo dos candidatos restantes.
  - Decida inclusão ou exclusão com justificativa.

- **Organize em planilha ou gerenciador de referências**:
  - Zotero, Mendeley, EndNote ou até uma tabela simples.
  - Registre: autor/ano, título, relevância (incluso/excluído e por quê).

### 5.4 Extrair informações e analisar criticamente

- Para cada estudo incluído, registre:
  - Autor, ano, país de origem.
  - Objetivo e pergunta de pesquisa.
  - Método (desenho, amostra ou dados, procedimentos).
  - Principais resultados e conclusões.
  - Limitações e possíveis vieses.

- **Avalie criticamente**:
  - Desenho metodológico: robusto? Apropriado para a pergunta?
  - Validade interna: possíveis fontes de erro ou viés?
  - Aplicabilidade: contexto, generalização para sua área.
  - Relevância para sua pesquisa: contribui diretamente ou oferece perspectiva interessante?

### 5.5 Sintetizar (não apenas resumir)

- **Agrupe e organize por temas, não por autor**:
  - Em vez de "Silva (2020) fez..., Costa (2021) fez...", organize: "Métodos causais para diagnóstico incluem X, Y, Z. Silva comparou X e Y com resultado..., enquanto Costa focou em Z com diferentes achados..."
  - Mostre convergências: onde os estudos apontam na mesma direção.
  - Mostre divergências: onde há conflito ou resultados contraditórios e por quê.
  - Identifique lacunas: o que ninguém estudou ainda? Que questões ficaram abertas?

- **Sempre conecte à sua pesquisa**: Como o estado da arte apoia, inspira ou contrasta com o que você quer fazer?

## 6. Tipos de revisão e rigor 

- **Revisão narrativa**: 
  - Mais descritiva e flexível.
  - Comum em TCCs, monografias e artigos introdutórios.
  - Busca pode ser menos sistemática, mas ainda deve seguir critérios mínimos.
  - Quando usar: quando o objetivo é mapear conceitos, debates e tendências gerais de um campo.

- **Revisão integrativa**: 
  - Combina estudos teóricos e empíricos para produzir uma síntese crítica mais ampla.
  - Mais rigorosa que narrativa; permite metodologias diversas.
  - Frequentemente usada em dissertações de mestrado.
  - Quando usar: quando se quer sintetizar diferentes tipos de estudo (qualitativos e quantitativos) e gerar uma compreensão integrada de um tema.

- **Revisão sistemática**: 
  - Segue protocolo explícito, com critérios de inclusão/exclusão predefinidos.
  - Busca reprodutível em múltiplas bases, registrada previamente (ex.: PROSPERO).
  - Frequentemente inclui síntese quantitativa (meta-análise).
  - Padrão-ouro em medicina e pesquisa com orientação de políticas públicas.
  - Quando usar: pergunta bem delimitada (ex.: "qual o efeito de X em Y?"), exigindo máxima rigor e transparência. Comum em doutorado e pesquisa avançada.

### Quadro comparativo:

| Aspecto | Narrativa | Integrativa | Sistemática |
|---------|-----------|-------------|------------|
| **Objetivo** | Mapear conceitos e debates | Sintetizar diversos tipos de estudo | Responder pergunta específica com evidência |
| **Sistematicidade** | Baixa a média | Média a alta | Muito alta |
| **Busca** | Flexível, com critérios mínimos | Estruturada, com critérios claros | Pré-registrada, reprodutível |
| **Quando usar** | TCC, artigos introdutórios | Dissertações de mestrado | Doutorado, pesquisa avançada, políticas |
| **Público-alvo** | Estudantes, leitores gerais | Pesquisadores, profissionais | Cientistas, formuladores de política |

## 7. Como saber se a revisão está boa? (Critérios de Qualidade)

### Checklist de qualidade:

- [ ] **Abrangência adequada**: cobriu as principais fontes e autores sem ser superficial ou excessivamente restrita?
- [ ] **Atualidade**: incluiu trabalhos recentes (últimos 2–3 anos) e clássicos relevantes?
- [ ] **Clareza de organização**: está bem estruturada (por temas, teorias, métodos ou cronologia) e é fácil acompanhar o raciocínio?
- [ ] **Capacidade crítica**: discutiu forças, fraquezas, lacunas e controvérsias, ou apenas descreveu?
- [ ] **Ligação com sua pesquisa**: fica claro por que isso importa para seu projeto? Ficou justificada a relevância de seu estudo?
- [ ] **Coerência argumentativa**: os parágrafos fluem logicamente, conectando ideias entre estudos?

### Rubrica de níveis:

- **Nível 1 (Iniciante)**: Revisão é principalmente uma lista descritiva de resumos de artigos. Pouca síntese ou análise crítica. Não fica claro como se conecta ao projeto.

- **Nível 2 (Intermediário)**: Revisão mostra alguma síntese e comparação entre estudos. Há início de análise crítica. Começa a conectar ao projeto, mas ainda faltam profundidade e clareza de lacunas.

- **Nível 3 (Avançado)**: Revisão é uma síntese crítica que integra estudos, identifica padrões, lacunas e controvérsias. Deixa clara a contribuição única de seu estudo. Argumentação é coerente e convincente.

Bancas e pareceristas olham para esses indicadores; melhor nota na rubrica = maior confiança no trabalho como um todo.

## 8. Erros comuns na Revisão da Literatura (e como corrigir)

### Erros principais:

1. **Listar estudos sem compará-los ou analisá-los**
   - Errado: "Silva (2020) fez X. Costa (2021) fez Y. Pereira (2022) fez Z."
   - Certo: "Três abordagens principais emergiram: X (Silva, 2020), Y (Costa, 2021) e Z (Pereira, 2022). Diferem em... A mais robusta é... porque..."

2. **Focar apenas no que foi feito, sem explicar por que isso importa**
   - Errado: Descrever métodos sem conectar à sua pergunta.
   - Certo: "Tal abordagem é relevante para meu estudo porque..."

3. **Não identificar lacunas ou contradições na pesquisa**
   - Errado: Assumir consenso onde há divergência; não notar o que não foi estudado.
   - Certo: "Enquanto alguns estudos indicam X, outros mostram Y. Essa divergência sugere que... Além disso, nenhum estudo explorou Z, que é relevante porque..."

4. **Não conectar as pesquisas anteriores com seu próprio estudo**
   - Errado: Revisar literatura como se fosse um exercício à parte.
   - Certo: "Meu trabalho se situa neste contexto: enquanto [lacuna], eu proponho [solução], o que pode contribuir porque..."

5. **Busca inadequada ou não transparente**
   - Errado: "Procurei no Google" ou não explicitar bases, termos, datas.
   - Certo: "Busquei em Scopus e Web of Science entre 2015 e 2025 com os termos..."

### Como corrigir:

**Para síntese e análise:**
- Leia e compare 2–3 estudos por temática, identificando: objetivo, método, resultados, limitações.
- Destaque padrões: "Três estudos concordam em... Um estudo discorda porque... A razão pode ser..."
- Use linguagem integradora: "Enquanto...", "Em contraste...", "Ampliando essa ideia...", "Isso deixa em aberto a questão...".

**Para capacidade crítica:**
- Não aceite nada como "verdade absoluta"; discuta: Qual era a amostra? Como foi medido? Há vieses conhecidos?
- Reconheça limitações: "Embora [estudo] sugira X, seu escopo foi limitado a [contexto], então é incerto se X vale para [seu contexto]."

**Para estrutura:**
- Escolha eixo claro: temas, teorias, métodos, ou evolução cronológica. Use seções e subseções (ex.: "2.1 Modelos causais em diagnóstico", "2.2 Modelos bayesianos").
- Cada parágrafo começa com ideia temática, inclui 1–3 estudos comparativos, e termina com ligação ao projeto.

**Para conexão com seu projeto:**
- No final da revisão, inclua um parágrafo (ou seção) tipo "Implicações para este estudo": como o que você leu justifica, motiva ou situa sua pergunta?

## 9. Ética e boas práticas na revisão bibliográfica

### Uso responsável de fontes:

- **Cite corretamente**: Use o padrão da sua instituição (ABNT, APA, Vancouver) e seja consistente.
- **Evite plágio**: Sempre atribua ideias e dados a seus autores. Parafrasear requer citação também.
- **Não distorça resultados**: Se um estudo concluiu X, mas seus dados apoiam melhor Y, respeite a conclusão original e discuta essa tensão, não omita.
- **Use sources primárias quando possível**: Não cite um estudo através de outro sem ter lido o original (exceto quando explícito: "conforme citado em [autor]").

### Transparência de busca e seleção:

- Mesmo em TCC, documente seu processo: que bases consultou, que termos usou, quantos artigos encontrou e quantos selecionou.
- Isso demonstra rigor e reduz o risco de viés de confirmação (tender a incluir só estudos que apoiam sua hipótese).
- Se perceber que sua busca é tendenciosa, ajuste critérios e documente a mudança.

### Responsabilidade na síntese:

- Quando há múltiplas interpretações possíveis, apresente as principais, não esconda controvérsias.
- Se um resultado é contra-intuitivo ou não apoiado por outros estudos, aponte isso; não o ignore para parecer mais simples.

***

### mensagem central e próximos passos

- **Mensagem central**: "Revisão bibliográfica não é burocracia: é a etapa que transforma uma ideia em pesquisa científica científica. Quanto melhor a revisão, mais clara a contribuição do seu trabalho e mais forte sua posição como pesquisador."

- **Próximos passos**: Depois de revisar, você estará pronto para:
  - Formular uma pergunta de pesquisa clara.
  - Escolher desenho e métodos bem justificados.
  - Escrever introdução e justificativa sólidas.

- **Mensagem de encerramento**: "Lembrem-se: nenhum pesquisador começa do zero. Vocês estão começando de onde a comunidade chegou. A revisão é como vocês pegam esse conhecimento coletivo, entendem, criticam e constroem sobre ele. Isso é ciência."

## 1. O que é **bibliometria?**
  - Entender o que é bibliometria e em que ela difere/complementa a revisão bibliográfica.  
  - Ver para que serve (mapear campos, tendências, autores, periódicos) e como pode aparecer em TCC/projetos em engenharia.  
- Perguntas para engajar  
  - “Como descobrir rapidamente quem são os principais autores e periódicos de um tema?”  
  - “Como saber se um tema está ‘em alta’ ou saturado na literatura?”  
- Conceito básico  
  - Bibliometria: uso de métodos quantitativos (contagem, estatística, redes) para analisar produção científica registrada (artigos, citações, periódicos, autores, instituições).  
  - Estudo bibliométrico como um tipo de pesquisa metodológica, distinto, por exemplo, de estudo de caso ou experimento.

***

## 2. O que a bibliometria analisa?

- Produtividade  
  - Quem mais publica? (autores, instituições, países)  
  - Quantos artigos saem por ano sobre o tema? (evolução temporal)  
- Impacto e circulação  
  - Citações recebidas, periódicos mais usados, índices (fale apenas de forma intuitiva: h-index, fator de impacto).  
- Estrutura e temas  
  - Coocorrência de palavras‑chave e áreas.  
  - Redes de coautoria, rede de citações (quem cita quem).  

***

## 3. Quando faz sentido usar bibliometria?

  - Usos típicos em engenharia e áreas aplicadas  
  - Mapear a evolução de um tema (por exemplo: “manutenção preditiva em redes elétricas”, “modelos de confiabilidade em sistemas mecânicos”).  
  - Identificar lacunas e oportunidades: subtemas pouco explorados, regiões/instituições pouco ativas.  
  - Apoiar decisões de onde submeter um artigo (periódicos mais relevantes na área).  
- Possíveis aplicações em TCC/projeto  
  - TCC puramente bibliométrico (ex.: “Análise bibliométrica da pesquisa sobre IoT em manufatura entre 2010–2025”).  
  - Módulo bibliométrico como parte da revisão: um pequeno mapeamento quantitativo para complementar a análise qualitativa.

***

## 4. Etapas de um estudo bibliométrico (versão simples)

1. Definição do tema e pergunta  
   - Ex.: “Como evoluiu a produção científica sobre controle preditivo aplicado a processos químicos nos últimos 15 anos?”  
   - Delimitar período, idioma, área, tipo de documento (artigos, conference papers etc.).  
2. Escolha de bases de dados  
   - Scopus, Web of Science, IEEE Xplore, SciELO, Google Scholar (com mais cuidado), entre outras.  
3. Estratégia de busca  
   - Palavras‑chave, combinações com operadores AND/OR, filtros por área, ano, tipo de documento.  
   - Registrar a estratégia (para reprodutibilidade).  
4. Limpeza e preparo dos dados  
   - Remover duplicados.  
   - Padronizar nomes de autores/instituições quando possível.  
   - Verificar consistência e campos disponíveis (título, resumo, palavras‑chave, ano, periódico, citações).  
5. Análises básicas  
   - Número de publicações por ano.  
   - Principais periódicos.  
   - Autores mais produtivos.  
   - Principais palavras‑chave e temas.  
6. (Opcional) Análises de redes  
   - Redes de coautoria.  
   - Redes de citações.  
   - Coocorrência de palavras‑chave.  

***

## 5. Exemplos de indicadores bibliométricos simples

  - Contagens e distribuições  
  - Número de artigos por ano → gráfico de linha (tendência).  
  - Artigos por país/instituição → gráfico de barras.  
  - Artigos por periódico → tabela com top N periódicos.  
- Palavras‑chave  
  - Frequência de palavras‑chave → identificar temas centrais.  
  - Evolução de termos ao longo do tempo (por exemplo, a entrada de termos como “machine learning” ou “IoT”).  
- Citações (bem intuitivo)  
  - Citações por artigo → identificar “clássicos” do tema.  
  - Média de citações por ano → sugerir impacto médio.


***

## 6. Limitações e cuidados

  - Qualidade dos dados  
  - Bases diferentes cobrem áreas diferentes; resultados dependem da base escolhida.  
  - Problemas com nomes de autores, homônimos, mudanças de filiação.  
  - Síndrome do “contar por contar”  
  - Fazer bibliometria sem boa pergunta acaba em gráfico sem interpretação.  
  - Risco de fetichizar números (fator de impacto, h-index) sem contexto.  
  - Ética e transparência  
  - Explicitar claramente critérios de busca, inclusão/exclusão, período.  
  - Não usar bibliometria para desqualificar pessoas/grupos de forma simplista.

## Modelo de **Capítulo de Revisão Bibliográfica**

Um modelo genérico pode ser organizado assim:

1. **Introdução ao tema e perguntas de pesquisa**  
   - Contextualização do fenômeno (ex.: transição energética, problema social/econômico).
   - Lacunas nas revisões anteriores e justificativa da nova revisão.
   - Formulação explícita das questões de pesquisa (RQ1, RQ2, RQ3…).

2. **Metodologia da revisão (*systematic literature review* (SLR) + bibliometria, se desejado)**  
   - Descrição do tipo de revisão (SLR, protocolo adotado, PRISMA (PRISMA é um 'Mapeador' Objeto-Relacional moderno e de código aberto para Node.js e TypeScript).
   - Bases de dados, período de busca, string de busca e uso de operadores booleanos/truncagem.
   - Critérios de inclusão/exclusão (tipo de documento, idioma, área, recorte temático).
   - Processo de filtragem (número inicial de estudos, etapas de filtragem, diagrama de fluxo tipo PRISMA).
   - Se pertinente, descrição da análise bibliométrica (software, indicadores, leis bibliométricas aplicadas).

3. **Resultados – visão geral bibliométrica**  
   - Evolução temporal das publicações e citações.
   - Principais periódicos/fontes, autores/redes de coautoria, acoplamento bibliográfico.
   - Mapa de palavras‑chave e clusters temáticos (com breve interpretação temporal dos focos).

4. **Resultados – análise de conteúdo**  
   - Definição de categorias analíticas centrais (no texto: tipo de impacto – econômico, ambiental, societal – e combinação simples/dupla/tripla).
   - Classificação dos estudos por abordagem metodológica (ex.: ABM/simulação, estudo de caso, modelo conceitual, modelo estatístico, survey, “outros”).
   - Discussão sintética de cada abordagem: para que é usada, vantagens/limitações, exemplos típicos.

5. **Taxonomia de variáveis (entrada/saída)**  
   - Extração manual das variáveis usadas nos modelos.
   - Classificação em macro‑categorias (econômicas, ambientais, societais, governança, conhecimento, tecnológicas).
   - Identificação de variáveis de entrada e de saída mais frequentes (eventualmente com tabelas).
   - Destaque de lacunas, como baixa padronização de variáveis societais.

6. **Relações entre metodologias, variáveis e impactos**  
   - Uso de diagramas (ex.: Sankey) ou tabelas para ligar tipos de modelos às variáveis de entrada/saída mais comuns.
   - Interpretação: quais metodologias capturam melhor múltiplos impactos, onde há concentração em certos indicadores etc.

7. **Discussão**  
   - Síntese interpretativa: o que os resultados dizem sobre o estado da arte e a maturidade do campo.
   - Destaque de ausência de quadro teórico consolidado, necessidade de integração interdisciplinar, complexidade do fenômeno.

8. **Conclusão e agenda de pesquisa**  
   - Recapitulação breve das principais contribuições (metodológicas e substantivas).
   - Proposição de agenda de pesquisa organizada em “temas abertos” e “trajetórias de pesquisa” para cada tema.
   - Limitações da revisão (ex.: janela temporal, bases, critérios) e indicações de extensões futuras.

## Modelo reutilizável para alunos

“template lógico” para qualquer revisão sistemática aplicada a um domínio técnico:

- **Seção 1 – Contexto e RQs**  
  “Descrever problema X, lacunas em revisões anteriores sobre X, justificar foco atual, derivar 2–4 RQs claras.”

- **Seção 2 – Protocolo da revisão**  
  “Especificar bases, período, string, critérios, etapas de triagem, número de estudos em cada etapa, diagrama PRISMA.”

- **Seção 3 – Panorama bibliométrico (opcional, mas desejável)**  
  “Apresentar evolução temporal, fontes/áreas dominantes, principais redes (autores, palavras‑chave), interpretar clusters.”

- **Seção 4 – Tipos de estudos e métodos**  
  “Definir categorias metodológicas adequadas ao domínio, classificar os artigos, discutir o papel de cada categoria.”

- **Seção 5 – Taxonomia de elementos centrais**  
  “Extrair e agrupar as principais ‘entidades’ do campo (variáveis, constructos, mecanismos) em macro‑categorias domínios‑específicas.”

- **Seção 6 – Mapeamento entre métodos e elementos**  
  “Mostrar como cada tipo de método combina com certos elementos (diagramas/tabelas) e que lacunas surgem dessas combinações.”

- **Seção 7 – Discussão crítica**  
  “Interpretar resultados à luz de teoria e prática, enfatizar complexidade, incerteza e necessidades de integração.”

- **Seção 8 – Conclusão + agenda**  
  “Enunciar contribuições, limitações e uma agenda estruturada em temas e trajetórias de pesquisa concretas.”
