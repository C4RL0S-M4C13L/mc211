**Objetivo da Aula**  
Capacitar alunos de graduação em Engenharia Elétrica a planejar, estruturar e iniciar um projeto de Trabalho de Conclusão de Curso (TCC), compreendendo as etapas fundamentais, requisitos formais e boas práticas acadêmicas.

---

## Estrutura Recomendada do Projeto de TCC

**1. Escolha do Tema e Definição do Orientador**
- Identifique uma área de interesse dentro da Engenharia Elétrica.
- Busque um professor orientador que atue ou tenha afinidade com o tema escolhido.
- A escolha do tema é crucial para o sucesso do projeto e deve considerar relevância, viabilidade e interesse pessoal[5][4][2].

**2. Elaboração do Plano de Trabalho (Pré-Projeto)**
- Descreva o tema do trabalho.
- Defina objetivos claros (geral e específicos).
- Proponha uma metodologia adequada ao tipo de projeto (experimental, estudo de caso, revisão bibliográfica, simulação, etc.).
- Liste referências bibliográficas iniciais.
- Monte um cronograma de execução com as principais etapas e prazos[5][1][3].

**3. Estrutura do Projeto de TCC**

| Seção                | Conteúdo Principal                                                                 |
|----------------------|-----------------------------------------------------------------------------------|
| Título e Autoria     | Defina um título claro e objetivo. Inclua nome do aluno e orientador.             |
| Introdução           | Apresente o tema, delimite o problema, justifique a escolha e defina objetivos.   |
| Fundamentação Teórica| Levante e discuta o estado da arte, conceitos e modelos teóricos relevantes.      |
| Metodologia          | Descreva detalhadamente os métodos, ferramentas e procedimentos a serem utilizados.|
| Cronograma           | Apresente um cronograma realista para o desenvolvimento do trabalho.              |
| Referências          | Liste as principais fontes utilizadas, seguindo normas ABNT ou IEEE.              |

No desenvolvimento do TCC (TCC II), o projeto evolui para incluir resultados, discussão, considerações finais e possíveis recomendações para trabalhos futuros[1][2][3].

---

## Etapas do Processo

**Planejamento**
- Escolha do tema e do orientador.
- Elaboração do pré-projeto/plano de trabalho.
- Levantamento bibliográfico inicial.

**Execução**
- Desenvolvimento do projeto conforme cronograma.
- Reuniões regulares com o orientador para acompanhamento e ajustes[4].

**Redação e Revisão**
- Escrita do texto conforme normas técnicas (ABNT ou IEEE).
- Revisão detalhada do texto: coerência, clareza e correção gramatical[3].
- Revisão por colegas e pelo orientador.

**Apresentação e Defesa**
- Apresentação oral do trabalho perante banca examinadora.
- Realização de eventuais correções sugeridas pela banca.
- Entrega da versão final e autorização para publicação[4][5].

---

## Boas Práticas e Dicas

- **Gestão do Tempo:** Divida o trabalho em etapas menores e estabeleça prazos para cada uma. Utilize cronogramas, planilhas ou aplicativos para acompanhar o progresso[3].
- **Rotina de Trabalho:** Mantenha uma rotina constante para evitar acúmulo de tarefas e estresse.
- **Normas Técnicas:** Atente-se às normas de formatação (ABNT/IEEE) para garantir padronização e credibilidade.
- **Discussão com Orientador:** Busque o orientador ativamente para esclarecimentos e feedbacks.
- **Originalidade e Ética:** Evite plágio e cite corretamente todas as fontes consultadas[3].
- **Conclusão:** Reforce os principais achados, indique limitações e sugira possíveis desdobramentos futuros.

## Exemplo de Cronograma Simplificado

| Etapa                        | Prazo Sugerido         |
|------------------------------|------------------------|
| Escolha do tema              | 1ª semana              |
| Definição do orientador      | 1ª-2ª semana           |
| Levantamento bibliográfico   | 2ª-4ª semana           |
| Elaboração do pré-projeto    | até 5ª semana          |
| Desenvolvimento do projeto   | 6ª-14ª semana          |
| Redação do texto             | 10ª-15ª semana         |
| Revisão e ajustes finais     | 15ª-16ª semana         |
| Defesa e entrega final       | 17ª-18ª semana         |

## Considerações Finais

O projeto de TCC é uma oportunidade de aplicar conhecimentos teóricos e práticos adquiridos ao longo do curso, desenvolver habilidades de pesquisa e comunicação científica, além de contribuir para sua formação como engenheiro. O sucesso depende de planejamento, disciplina e interação produtiva com o orientador[1][2][3][4][5].
