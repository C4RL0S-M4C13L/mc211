- Padronização (ABNT, IEEE, etc.)

## Estrutura Básica de um Documento LaTeX

- Instalação e escolha do editor (Texmaker, Overleaf, VSCode)
- Estrutura mínima:
  - Preambulo (`\documentclass`, `\usepackage`)
  - Corpo (`\begin{document} ... \end{document}`)
- Classes principais: `article`, `report`, `book`
- Pacotes essenciais: `amsmath`, `graphicx`, `babel`, etc.

---



## Exemplo de Documento Básico

````
\documentclass[a4paper,12pt]{article}
\usepackage[brazil]{babel}
\usepackage[utf8]{inputenc}
\usepackage{amsmath,amssymb}

\begin{document}
Olá, mundo!

\end{document}
````

# Relatório de Circuitos Elétricos

## Introdução

Este relatório apresenta a análise de um circuito RC.

## Materiais

| Componente | Valor |
|------------|-------|
| Resistor   | 1kΩ   |
| Capacitor  | 10uF  |

## Esquemático

![Circuito RC](https://upload.wikimedia.org/wikipedia/commons/e/e0/1st_Order_Lowpass_Filter_RC.svg)

## Equação

A resposta do circuito é dada por:

$$
V(t) = V_0 \cdot e^{-\frac{t}{RC}}
$$


## Código de Simulação

````
import numpy as np
import matplotlib.pyplot as plt

R = 1000  # Ohms
C = 10e-6 # Farads
t = np.linspace(0, 0.1, 1000)
V0 = 5
Vt = V0 * np.exp(-t/(R*C))

plt.plot(t, Vt)
plt.xlabel('Tempo (s)')
plt.ylabel('Tensão (V)')
plt.title('Resposta do Circuito RC')
plt.show()
````
