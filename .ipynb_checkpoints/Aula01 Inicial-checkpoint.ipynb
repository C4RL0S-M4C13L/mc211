# EMENTA
Disciplina de caráter teórico e motivacional que visa apresentar aos estudantes uma visão global da ciência e sua evolução, fomentando uma análise crítica e reflexiva do papel da ciência e do engenheiro na sociedade, além de apresentar as possibilidades dentro da carreira e do curso. A disciplina contribuirá também no desenvolvimento de competências como comunicação, ética profissional e responsabilidade, compreensão da importância de ser um agente preservador do meio ambiente e o desenvolvimento do pensamento crítico

---
# OBJETIVOS
- Assegurar que os alunos tenham uma visão global sobre a evolução da ciência.
- Oferecer aos alunos a possibilidade de analisar criticamente o papel da ciência e seus métodos.
- Assegurar que os alunos tenham conhecimento da evolução da Engenharia Elétrica ao longo do tempo.
- Assegurar que os alunos tenham consciência do papel social do engenheiro como vetor do desenvolvimento nacional.
- Oferecer aos alunos a possibilidade de um contato com as atividades corriqueiras ligadas à carreira e ao curso de engenharia elétrica

---
# METODOLOGIA DO ENSINO
Aulas expositivas podendo ser incorporado metodologias ativas. Neste contexto, os discentes são estimulados a trabalhar nas atividades propostas visando a articulação entre as competências e habilidades a serem desenvolvidas na disciplina, cujo desempenho é verificado em avaliações das atividades propostas, relatórios e provas.
