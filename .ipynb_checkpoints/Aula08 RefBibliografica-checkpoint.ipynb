## 1. Introdução: Por que Referenciar? 

### Discussão Inicial
- **Pergunta para a turma:**  
  “Por que vocês acham que é importante citar as fontes em um trabalho científico?”

## Uma referência tem, pelo menos, três finalidades:

1. Dar o devido crédito ao autor do texto original ao qual se faz referência. É uma questão de honestidade intelectual.

2. Possibilitar ao leitor a localização da fonte de onde foi extraída a informação. Ele pode querer ir até lá buscar mais detalhes sobre o tema. Dotar o autor de uma "memória auxiliar". Mais tarde talvez ele queira voltar à fonte. 

3. Além disso, a citação de uma fonte reconhecidamente confiável pode dar maior credibilidade ao que o autor escreve em razão da autoridade da fonte citada.

### Pontos-Chave
- **Credibilidade:** Demonstra que o trabalho está fundamentado em conhecimento reconhecido.
- **Ética e Plágio:** Evita apropriação indevida de ideias e garante honestidade intelectual.
- **Rastreabilidade:** Permite que outros encontrem e verifiquem as fontes.
- **Avanço científico:** Facilita o aprofundamento e a continuidade das pesquisas.




---

## 2. O que é uma Referência Bibliográfica?

- **Definição:**  
  Conjunto de informações que identifica e localiza uma fonte utilizada em um trabalho.
- **Diferença entre referência e citação:**  
  - *Citação* é a menção no texto.  
  - *Referência* é a lista detalhada ao final do trabalho.



---

## 3. Normas e Estrutura das Referências 

### Normas Utilizadas
- **ABNT NBR 6023:2018** (principal no Brasil)
- Outras normas (APA, IEEE) – mencionar que IEEE é comum em Engenharia, mas a ABNT é padrão acadêmico nacional.

### Elementos Essenciais (ABNT)
- **Autor(es)**
- **Título**
- **Edição**
- **Local**
- **Editora**
- **Ano**
- **Para fontes online:** URL e data de acesso



### Exemplos Práticos

**Livro:**
```
SOBRENOME, Nome. Título: subtítulo. Edição. Local: Editora, ano.
```
**Artigo:**
```
SOBRENOME, Nome. Título do artigo. Título da Revista, volume, número, páginas, ano.
```
**Site:**
```
AUTOR(ES). Título. Disponível em: . Acesso em: dia mês ano.
```

## 4. Como Fazer Referências na Prática

### Passo a Passo
1. **Identifique o tipo de fonte** (livro, artigo, site, norma técnica, etc.).
2. **Recolha todos os dados necessários**.
3. **Siga a ordem e a pontuação da norma ABNT**.
4. **Liste as referências em ordem alfabética** pelo sobrenome do autor.

### Exercício Rápido (em duplas)
- Cada dupla recebe uma fonte (livro, artigo ou site) e monta a referência conforme a ABNT.
- Compartilhar 2-3 exemplos com a turma.



---

## 5. Boas Práticas e Erros Comuns

### Boas Práticas
- Conferir sempre os dados na fonte original.
- Manter o padrão de formatação.
- Atualizar URLs e datas de acesso para fontes online.
- Referenciar apenas o que foi realmente consultado e citado.


### Erros Comuns
- Omissão de elementos essenciais (ex: ano, editora).
- Mistura de normas diferentes.
- Citar obras não consultadas.
