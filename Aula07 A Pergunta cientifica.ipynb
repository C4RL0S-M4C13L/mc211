## 3. O que é uma pergunta científica?
1. “Incerteza sobre algo na população em questão que o investigador quer resolvê-la através da realização de medidas nos sujeitos do
estudo”
2. Situação não resolvida, objeto de investigação/discussão, em qualquer ramo da ciência

## 4. Escolhendo um tópico - exemplos médicos!! Como seriam na engenharia?
1. A ideia pode vir de experiências de vida
2. O que eu tenho? (diagnóstico)
3. Quão ruim isto é? (prognóstico)
4. Posso transmitir para minha família? (historia natural)
5. Como eu peguei isso? (etiologia)
6. Isto vai me ajudar a me curar? (tratamento)

## 5. Escolhendo um tópico
1. Deve conter um método validado de medição de desfecho (resultado)
2. Exemplos mensuráveis:
3. Medindo gravidade da doença
4. Desenvolvimento de medidas de desfecho final (nível de hipoacusia, valor de acuidade visual, espessura de estruturas anatômicas, etc)

## 6. Características de uma boa pergunta científica
1. Factibilidade
2. Interesse científico
3. Inédita, inovadora
4. Ética
5. Relevante

### 7. Factibilidade
1. Número adequado de sujeitos da pesquisa
2. Adequado conhecimento técnico
3. Pode ser realizada com o tempo e recursos disponíveis
4. Mensurável, com foco
5. Numero mínimo de sujeitos que devem ser recrutados para que estatisticamente possamos responder a pergunta elaborada
6. Quanto mais prevalente é a doença pesquisada, ou maior é a diferença estimada entre os grupos, menor é o número de pessoas que precisam ser incluídas
7. Estudo “piloto”
8. Adequado conhecimento técnico:
9. Habilidade, equipamentos e experiência necessária para planejamento, recrutamento, medição de variáveis estudadas, manejo e análise de dados
10. Colaboradores/consultores
11. Pode ser realizada com o tempo e recursos disponíveis:
12. Estimativa de custos
13. Realizar ajustes ou até cancelar o estudo na ausência de tempo e recursos
14. Mensurável, com foco:
15. Só estudar o que você consegue medir
16. Foco em uma (ou poucas) pergunta (s) sobre um assunto
17. Outras perguntas relevantes sobre o mesmo assunto podem ser respondidas em estudos futuros

#### 8. Interesse científico
1. Obter a resposta intriga o pesquisador e seu grupo:
2. Motivações: financeiras, para crescimento profissional, pelo prazer em descobrir algo novo


#### 9. Inédita, inovadora
1. Confirma, refuta ou estende o conhecimento prévio
2. Fornece novo conhecimento
3. Pesquisa não-inédita: pode ser testada em diferentes populações
4. Como saber?
5. Revisão de literatura
6. Busca em listas de projetos financiados (FAPESP, CNPq)


#### 10. Ética
1. Que será aprovada por um comitê de ética
2. Que não coloque os sujeitos da pesquisa em risco

### Como identificar lacunas no conhecimento científico?
1. Primeiro precisamos identificar qual é o conhecimento científico no
assunto a ser estudado
2. Boa atualização sobre o tema, em boas referências
3. Referências mais recentemente publicadas possível
