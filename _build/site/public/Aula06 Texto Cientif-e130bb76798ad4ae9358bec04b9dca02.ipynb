## O que é Tese/Dissertação: 

* **Tese** é um assunto, um tema, um objeto. É uma proposição que se apresenta para ser discutida e defendida por alguém, com base em determinadas hipóteses ou pressupostos. Do grego “thesis” que significa “proposição”.  A expressão “em tese” significa “de modo geral”, “conforme o que se supõe”, “em princípio”, “em teoria”.

* **Tese de doutorado** é um trabalho acadêmico formulado nas escolas superiores para ser defendido em público, para um grupo de doutores, como condição para obtenção do título de doutorado. A tese acadêmica é uma abordagem de um único tema, resultado de pesquisa na área em que se situa,   elaborada seguindo metodologia específica. A pesquisa para a tese acadêmica pode ser teórica, de campo, documental, experimental, histórica ou filosófica. Deve colocar e solucionar um problema demonstrando hipóteses formuladas em  evidências de fatos sempre obedecendo a um raciocínio lógico. A tese de doutorado deve apresentar um progresso para a área científica em que está situada. “Tese” é a publicação de trabalho acadêmico.

Segundo a ABNT (NBR 6022, 2003, p.2), o **artigo científico** pode ser definido como a “publicação com autoria declarada, que apresenta e discute ideias, métodos, técnicas, processos e resultados nas diversas áreas do conhecimento”.

O artigo científico caracteriza-se por um texto científico cuja função é relatar os resultados, sendo esses calçados de **originalidade**, provenientes de uma *dada pesquisa*. Dessa maneira, ele, materializado sob a forma de um **relato acerca dos resultados originais de um estudo realizado**, torna-se publicamente conhecido por meio de revistas científicas. 

Santos (2007, p. 43), **“são geralmente utilizados como publicações em revistas especializadas, a fim de divulgar conhecimentos, de comunicar resultados ou novidades a respeito de um assunto, ou ainda de contestar, refutar ou apresentar outras soluções de uma situação convertida”**.

## Texto científico: elementos textuais

Geralmente, os textos científicos são compostos por título, resumo, introdução, teoria, materiais e métodos, resultados e discussões, conclusão e referências. Cada elemento tem a sua importância e deve contemplar os aspectos descritos a seguir.

1. Título

A função do título é atrair a atenção do leitor. Use palavras específicas associadas ao seu trabalho. Uma boa estratégia, sempre que possível, é usar palavras-chave que enfatizem sua contribuição. Além disso, é preciso ser conciso para que seu título não contenha termos que não traz nenhuma informação.

2. Resumo

O resumo deve abordar todos os aspectos principais do texto. Um resumo bem escrito deve ser conciso, permitindo ao leitor identificar rapidamente as principais contribuições do artigo para a área. O resumo deve começar com uma contextualização geral do tema, depois deve apresentar uma lacuna não resolvida na literatura, em seguida deve falar a proposição do artigo, depois deve falar sobre a metodologia, sobre os principais resultados e fechar com as conclusões. Um bom exercício para melhorar a elaboração de resumo é selecionar bons artigos e identificar qual parte do resumo se refere a cada elemento (contextualização, gap, contribuição, metodologia, resultados e conclusões). Fique atento ao limite de palavras estabelecido por cada revista.

Divida o texto em 5 partes:

    a. Adicione uma frase introdutória de impacto com um máximo de 2 linhas;

    b. Descreva claramente os objetivos e os tratamentos experimentais em um máximo de 4 linhas;

    c. Descreva resumidamente a metodologia do trabalho em um máximo de 4 linhas;

    d. Descreva os seus resultados, utilize números e percentagens para facilitar o entendimento. Utilize o máximo de linhas possíveis;

    e. Finalize com uma frase indicando como os seus resultados contribuem para a área de pesquisa/conhecimento. 

3. Introdução

Esse é o ponto mais importante do trabalho. É nele que o leitor vai se motivar em continuar ou não na leitura do seu texto. As principais sub-seções que compõem a introdução são:

    a. Contextualização e gap: Apresente o campo de pesquisa e mostre a importância da área principal. Familiarize os termos e processos. 
    b. Deve introduzir a ideia do trabalho e como ele está relacionado com os demais trabalhos apresentados na literatura. Um cuidado adicional é que o problema a ser estudado deve ter sido indicado na revisão da literatura (revisão bibliográfica → Conferir sempre no Web of Science). Não somos nós que dizemos que é um problema. Portanto, nesta parte o texto deve deixar claro qual a importância da área e o que ainda não foi feito ou resolvido.
    c. Tese: considerando os trabalhos apresentados na literatura, como se situa a proposta do autor? Qual a proposição (objetivos) que o autor apresenta para ser defendida no final do trabalho? Porque esse estudo é importante?
    d. Agenda: para atingir os objetivos acima, como foi estruturado o trabalho? O que se espera em cada seção?

Na introdução se preocupe com a estrutura do texto. Não faça parágrafos com grande desigualdade de extensão. Uma boa referência para esse texto é [link](http://sepwww.stanford.edu/sep/prof/Intro.html)

4. Teoria

Deve apresentar os principais modelos ou equações em que se baseia o trabalho. Cuidado, não seja prolixo. Apresente apenas o que seria importante para o entendimento do trabalho.

5. Materiais e Métodos

Descreva os seus algoritmos, casos de teste dos algoritmos e dados utilizados no trabalho. Quando descrever os seus algoritmos use as ferramentas adequadas como as utilizadas em eng. de software. Descreva em linhas gerais o seu ambiente computacional (SO, CPU etc) e passe uma ideia ao leitor do tempo computacional de execução do seu software. Para descrever o método, use fluxograma ou pseudo código. Para elaboração de fluxogramas, o Mathcha permite fazer a esboço online, exportar o código e formar a figura do fluxograma diretamente no latex utilizando o TikZ. Para elaboração do pseudocódigo é possível utilizar o latex diretamente. Veja o exemplo de pseudocódigo no final do texto.

6. Resultados

Apresente os resultados na mesma ordem que foram descritos na seção Materiais e Métodos. Certifique-se através da literatura quais as métricas são adequadas para avaliar os resultados. Termine com uma tabela consolidando os resultados apresentados.

7. Discussão e Conclusão

    a. Discussão: reintrodução do seu contexto do trabalho e seus objetivos. Compare os seus resultados com os descritos na literatura. A discussão deve ser totalmente embasada nos seus resultados e na literatura.
    b. Contribuições do trabalho:
    c. Conclusão: dê a sua opinião, faça o fechamento do trabalho e indique os próximos passos.

8. Referências: 

Todas as informações ou ideias devem ser referenciadas, mesmo que sejam de outros trabalhos de sua autoria. Nunca transcreva trechos de outros textos, isso é plágio. Escreva com suas próprias palavras e referencie.

Existem diferentes formas de mostrar as citações ao longo do texto e para organizar a lista de trabalhos citados. Cada revista tem o seu padrão. O ideal é consultar o guia para autores da revista de interesse e usar o modelo disponibilizado pela revista. Para dissertação e tese, consulte o modelo do seu programa de pós-graduação.

Procure citar somente trabalhos dos últimos 4 anos, com exceção aos trabalhos antigos que são grandes nomes da área.


Ao longo do texto outros cuidados são fundamentais:

1. Capriche nas figuras. O tamanho da fonte utilizado nos eixos é grande o suficiente para ser possível ler? A figura é de fácil entendimento? Você adotou cores adequadas para melhorar a visualização das informações contidas na figura? E outros cuidados, até que a figura fique ótima.

2. Se você colocou uma figura no seu texto, ela deve ser discutida no mesmo. As legendas devem ser grande o suficiente para que o leitor entenda o que se deseja com ela.

3. Cuidado redobrado com a bibliografia. A pior coisa para quem lê o trabalho é procurar uma referência que não existem.

4. Seu artigo é reprodutível? Tenha certeza de que sua metodologia e resultados foram descritos na profundidade necessária para que o leitor possa reproduzi-lo. Veja mais sobre reprodutibilidade em 
[link](http://arstechnica.com/science/news/2012/02/science-code-should-be-open-source-according-to-editorial.ars e em https://www.ncbi.nlm.nih.gov/pmc/articles/PMC2981311/?tool=pubmed .)

5. e o seu texto for um artigo, é necessário elaborar uma carta de encaminhamento.a carta de encaminhamento deve apresentar ao editor que a revista já publica na linha do trabalho submetido e como esse trabalho se encaixa nesses trabalhos. De preferência cobrindo alguma lacuna deixada pela sucessão de artigos anteriores. O seu texto para ser publicado deve primeiramente agradar ao editor e depois os revisores. Caso contrário o artigo nem é revisado.

6. O texto submetido deve, na média, atender os requisitos gráficos (numero de parágrafos por seção, quantidade de figuras, numero de páginas, etc.) da revista pretendida.

7. Uma boa referência é: https://www.elsevier.com/journals/international-journal-of-fatigue/0142-1123/how-to-write-an-article

Portanto, existe um padrão a ser seguido durante a escrita de artigo científico. Bons trabalhos podem ser recusados se não conter todos os elementos de um bom texto científico. Confira no anexo um checklist a ser verificado antes da submissão do artigo.



## Tipos de Gaps em Trabalhos Científicos

**Gap de pesquisa** refere-se a uma lacuna de conhecimento, ou seja, aspectos de um tema que ainda não foram suficientemente explorados, compreendidos ou resolvidos pela literatura científica existente[1][2]. Identificar e explicitar o gap é fundamental para justificar a relevância de um trabalho científico, pois demonstra como a pesquisa proposta pode contribuir para o avanço do conhecimento.

### Principais Tipos de Gaps em Pesquisa Científica

Embora a literatura nem sempre padronize uma tipologia única, os gaps mais comuns em trabalhos científicos podem ser classificados da seguinte forma:

- **Gap de Conhecimento**: Refere-se à ausência de informações, dados ou explicações sobre determinado fenômeno. É o tipo mais clássico, quando simplesmente não há estudos suficientes sobre um tema específico[1][2].

- **Gap Metodológico**: Ocorre quando há limitações, deficiências ou ausência de determinadas abordagens metodológicas nos estudos existentes. Por exemplo, predominância de métodos qualitativos quando seriam necessários estudos quantitativos, ou vice-versa[1].

- **Gap Teórico**: Relaciona-se à falta de desenvolvimento, integração ou aplicação de teorias para explicar determinado fenômeno. Pode envolver a ausência de modelos teóricos robustos ou a necessidade de revisão de conceitos existentes[1][2].

- **Gap Prático ou Aplicado**: Surge quando há carência de estudos que abordem a aplicação prática do conhecimento produzido, ou quando resultados de pesquisas não são testados em contextos reais[1].

- **Gap Populacional ou Contextual**: Refere-se à ausência de estudos em determinados grupos populacionais, regiões geográficas, setores ou contextos específicos. Por exemplo, pesquisas concentradas em países desenvolvidos, mas escassas em países em desenvolvimento[1][2].

- **Gap Temporal**: Ocorre quando há desatualização dos estudos, ou seja, o tema foi pesquisado no passado, mas não há investigações recentes que considerem mudanças tecnológicas, sociais ou econômicas[1].

### Como Identificar Gaps

- Leitura crítica e sistemática da literatura científica (estado da arte)[1].
- Análise das seções de "limitações" e "pesquisas futuras" em artigos relevantes[1].
- Observação de contradições, inconsistências ou resultados inconclusivos em estudos anteriores[1][2].

### Resumo dos Tipos de Gaps

| Tipo de Gap           | Descrição                                                                 |
|-----------------------|---------------------------------------------------------------------------|
| Gap de Conhecimento   | Falta de estudos ou informações sobre o tema                              |
| Gap Metodológico      | Limitações ou ausência de abordagens/metodologias adequadas               |
| Gap Teórico           | Falta de modelos, teorias ou integração teórica                           |
| Gap Prático/Aplicado  | Carência de aplicação prática ou validação em contextos reais             |
| Gap Populacional      | Ausência de estudos em determinados grupos, regiões ou contextos           |
| Gap Temporal          | Desatualização dos estudos existentes                                     |

Esses gaps podem aparecer isoladamente ou combinados, e a identificação clara deles é essencial para fundamentar a originalidade e a relevância de qualquer trabalho científico[1][2].

## Checklist para submissão

1. [ ] O título é conciso e atrativo. Além disso, ele contém palavras-chaves que enfatizam o tema e a contribuição. 

2. [ ] O resumo permite que o leitor identifique rapidamente as principais contribuições do artigo para a área. É possível identificar no resumo a contextualização, o gap, a contribuição, a metodologia, os resultados e as conclusões. 

3. [ ] A introdução motiva o leitor a ler o restante do texto. A contextualização e gap ficaram claros e foram definidos a partir da revisão da literatura. O propósito do artigo foi definido a partir do gap e apresentado de forma clara, mostrando a importância da contribuição para a área. Por fim, a agenda mostra como o texto está estruturado para atingir os objetivos e o que se espera em cada seção. 

4. [ ] Na teoria, foram apresentados todos os modelos e equações importantes para o entendimento e reprodução do trabalho. 

5. [ ] Em materiais e métodos, os algoritmos, casos de teste e dados utilizados foram descritos e todas as escolhas foram justificadas. Para facilitar o entendimento, foi utilizado fluxogramas ou pseudocódigos que dão a profundidade necessária para que o leitor possa reproduzir o trabalho. 

6. [ ] O ambiente computacional foi descrito em linhas gerais (SO, CPU, etc.) e o tempo computacional para execução também. 

7. [ ] Os resultados foram apresentados na mesma ordem que foram descritos na seção de Materiais e Métodos. Todas as métricas de avaliação são adequadas e por fim foi apresentado uma tabela consolidando os resultados. 

8. [ ] Meu texto não possui plágio.

9. [ ] Todas as referências foram verificadas quanto a existência e adequação ao modelo da revista. Além disso, foram utilizadas referências dos últimos 4 anos, com exceção aos trabalhos antigos que são grandes nomes da área. 

10. [ ] As figuras estão adequadas. São legíveis e favorecem a interpretação de todas as informações contidas em cada uma delas. Além disso, as legendas são grandes o suficiente para que o leitor entenda a importância de cada figura. 

11. [ ] A carta de encaminhamento diz ao editor como o trabalho se encaixa na revista. Para isso, foi apresentado que a revista já publica na linha do trabalho e qual a lacuna deixada pelos artigos anteriormente publicados na revista que estamos propondo resolver.
