## 1. Objetivos da aula

Ao final da aula, o aluno deve ser capaz de:  
- Formular hipóteses claras e testáveis.  
- Definir variáveis e indicadores coerentes com as hipóteses.  
- Planejar a coleta de dados (quando, onde, com quem, como).  
- Reconhecer fontes básicas de erro e viés no estudo.  

***

## 2. Desenho do estudo: visão geral

Definição simples: **“Desenho do estudo é o plano lógico que liga a pergunta de pesquisa aos dados que serão coletados e à forma de analisá‑los.”**

Elementos centrais a enfatizar (independente de ser quantitativo ou qualitativo):  
- **Pergunta de pesquisa / objetivo**  
- **Hipóteses (quando fizer sentido)**  
- **Unidade de análise** (pessoa, empresa, hospital, município, etc.)  
- **População e amostra**  
- **Variáveis / indicadores**  
- **Estratégia de coleta** (instrumentos, fontes de dados, cronograma)  

p. ex., “fatores associados à evasão em disciplinas de cálculo”, “satisfação de pacientes em um ambulatório”, “impacto de um curso de capacitação na produtividade”

***

## 3. Hipóteses, variáveis e indicadores

### 3.1 Hipóteses

Definição operativa:  
- Hipótese é uma proposição sobre a relação entre variáveis que pode ser confrontada com dados.  
- Idealmente é **específica**, **direcional (quando couber)** e **falsificável**.

Estrutura típica:  
- “Se X aumenta/diminui, então Y tende a aumentar/diminuir.”  
- “Alunos que participam do programa A terão, em média, maior desempenho que alunos que não participam.”

Recomendação:  
- Separar **hipótese substantiva** (teórica) da **hipótese estatística** (H0, H1).  
- Evitar hipóteses vagas (“tecnologia influencia aprendizagem”).

### 3.2 Variáveis

Classificações:

- **Quanto ao papel no modelo**  
  - Independente (preditor, exposição, causa presumida).  
  - Dependente (desfecho, resposta).  
  - De controle / confundidora (para ajustar, reduzir viés).  
  - Moderadora (altera a força ou direção da relação X→Y).  
  - Mediadora (mecanismo intermediário entre X e Y).

- **Quanto ao tipo de medida**  
  - Nominal (categorias sem ordem: sexo biológico, área de atuação).  
  - Ordinal (categorias ordenáveis: “discordo/…/concordo”).  
  - Intervalar (diferenças fazem sentido; sem zero absoluto clássico: temperatura em Celsius).  
  - Razão (possui zero verdadeiro: renda, idade, número de consultas).

Recomendação: sempre especifique **unidade** e **domínio** da variável (ex.: “idade em anos completos, 18–80 anos”).

### 3.3 Indicadores

Ideia-chave:  
- Indicador é **a forma concreta de observar/medir** uma variável mais abstrata.

Exemplos:  
- Variável “desempenho acadêmico” → indicadores: média semestral, aprovação/reprovação em disciplina‑chave, nota em prova padronizada.  
- Variável “adesão ao tratamento” → indicadores: % doses tomadas, contagem de comprimidos, registro em prontuário, resposta a questionário.

Critérios para bons indicadores:  
- **Validade**: o indicador realmente representa o construto?  
- **Confiabilidade**: a medida é estável e consistente?  
- **Factibilidade**: é possível coletar com os recursos e tempo disponíveis?

Exemplo:  
- Dar 2–3 variáveis abstratas (“qualidade de vida”, “engajamento em aula”, “clima organizacional”) (e proponha indicadores mensuráveis).

## 4. Planejamento da coleta de dados


### 4.1 De quem? (população e amostra)

- **População‑alvo**: conjunto de interesse teórico (ex.: todos os alunos de graduação de uma universidade).  
- **População acessível**: quem você realmente consegue alcançar (ex.: alunos matriculados em 2025‑1 em 3 cursos).  
- **Amostragem** (noções básicas):  
  - Probabilística (aleatória simples, estratificada, por conglomerados).  
  - Não probabilística (conveniência, intencional, bola de neve).

Reforço conceitual:  
- Quanto mais não probabilística a amostragem, maior o cuidado ao generalizar resultados.

### 4.2 O que e como medir? (instrumentos)

- Questionários (online/papel).  
- Entrevistas (estruturadas, semiestruturadas).  
- Observação (estruturada, participante).  
- Dados secundários (registros, bases administrativas, prontuários, sistemas institucionais).  
- Medidas objetivas (testes, sensores, logs de sistemas, dados de uso de software).

Boas práticas:  
- Usar instrumentos **validados** quando houver (por exemplo, escalas consagradas).  
- Fazer **pré‑teste/piloto** para verificar compreensão, tempo, problemas de registro.

### 4.3 Quando e onde? (logística)

- Momento da coleta (antes/depois de intervenção; ponto único; seguimento em coortes).  
- Local (sala de aula, laboratório, ambiente on‑line, campo).  
- Duração estimada e janelas de coleta.  
- Quem aplica/coleta (pesquisador, equipe treinada, automatizado).

### 4.4 Questões éticas

Mesmo que seja só uma lâmina:  
- Consentimento livre e esclarecido.  
- Privacidade, anonimização ou pseudoanonimização.  
- Minimização de dano/risco para participantes.

## 5. Noções de erros e vieses

Use linguagem simples e exemplos próximos da realidade dos alunos.

### 5.1 Erros aleatórios

- Flutuações imprevisíveis nas medidas, ruído.  
- Causas: variação de humor do respondente, condições ambientais, precisão limitada do instrumento.  
- Efeito: afetam a **precisão** (aumentam variância, intervalo de confiança mais amplo).  
- Estratégias de mitigação: padronizar procedimentos, treinar coletores, aumentar tamanho de amostra.

Metáfora útil: “tiros em um alvo”: erro aleatório → tiros espalhados, mas sem direção preferencial.

### 5.2 Erros sistemáticos (vieses)

Definição simples:  
“Viés é um erro consistente, que desloca as estimativas numa direção específica, fazendo o estudo tender a conclusões erradas.”

Tipos que valem a pena comentar:

- **Viés de seleção**  
  - Quando a amostra não representa a população‑alvo.  
  - Exemplos: só responde o questionário quem está muito motivado; excluir sistematicamente pessoas com menos acesso à internet.  
  - Consequência: distorção na estimativa de frequências ou associações.

- **Viés de informação (medida)**  
  - Erros sistemáticos na forma de obter informações (classificação errada, medidas tendenciosas).  
  - Exemplos: perguntas mal formuladas, instrumento com viés de desejabilidade social (responder o socialmente aceitável), registro incompleto em prontuários.

- **Viés de confundimento**   
  - Uma terceira variável relacionada à exposição e ao desfecho que não está bem controlada.  
  - Exemplo didático: associação entre “uso de óculos” e “nota alta” pode ser confundida por “horas de estudo”.

- **Viés de não resposta / perda de seguimento**  
  - Quando quem não responde ou abandona o estudo é sistematicamente diferente de quem permanece.

Estratégias gerais de mitigação:  
- Planejar amostragem mais próxima do aleatório possível.  
- Definir e padronizar claramente variáveis, indicadores e procedimentos de coleta.  
- Usar instrumentos testados e fazer piloto.  
- Documentar perdas, recusas e faltas de dado; sempre discutir essas limitações na análise.
